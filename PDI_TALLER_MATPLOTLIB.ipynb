#### 1) Funci√≥n seno y coseno

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10,5))  #modificar el tama√±o (x,y)
plt.plot(x,y_sin,'g',label='$f(x)=sin(x)$')
plt.plot(x,y_cos,'b',label='$f(x)=cos(x)$')
plt.xlabel('x')  #etiqueta en el eje x
plt.ylabel('y') #etiqueta eje y
plt.title(f'Gr√°fica de Funciones $sin(x)$ y $cos(x)$')
plt.legend()
plt.show()
```

#### 2) Ajustar los limites de los ejes: x(0,10) y y(-10,100)

```python
x = np.linspace(-10,10,100)
y = x**2

plt.figure(figsize=(10,5))  #modificar el tama√±o (x,y)
plt.plot(x,y,'g',label='$f(x)=x^2$')
plt.xlabel('x')  #etiqueta en el eje x
plt.ylim(-10,100)
plt.xlim(0,10)
plt.ylabel('y') #etiqueta eje y
plt.title(f'Gr√°fica')
plt.legend()
plt.show()
```

#### 3) Agregar anotaciones

```python
x = np.linspace(-10,10,100)
y_th = np.tanh(x)

plt.figure(figsize=(10,5))  
plt.plot(x,y_th,'g',label='$f(x)=tanh(x)$')
plt.annotate("Punto Medio $f(0)$", xy=(0,0), xytext=(-0.5,0.5), arrowprops=dict(facecolor='red', shrink=0.05))
plt.show()
```

#### 4) Subtramas y diferentes tipos de datos

```python
x = np.linspace(0,10,100)
y_l = x**3+1

y_b = np.random.randint(1, 10, 4)
etiquetas = ['A', 'B', 'C', 'D', 'E']

# Datos para el gr√°fico de pastel
sizes = [15, 30, 45, 10]
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']

# Crear una figura y un arreglo de ejes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Gr√°fico de l√≠neas
axs[0].plot(x, y_l, marker='o')
axs[0].set_title('Gr√°fico de L√≠neas')

# Gr√°fico de barras
axs[1].bar(labels, y_b, color='orange')
axs[1].set_title('Gr√°fico de Barras')

# Gr√°fico de pastel
axs[2].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[2].set_aspect('equal')  # Para que el pastel sea circular
axs[2].set_title('Gr√°fico de Pastel')

# Ajustar layout
plt.tight_layout()

plt.show()
```

#### 5 Mapas de Calor

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

np.random.seed(0)
matrix = np.random.rand(4,4) #Matrix 4x4 de valores aleatorios

matrix_interpo = gaussian_filter(matrix, sigma=1)

plt.figure(figsize=(8, 6))
plt.imshow(matrix_interpo, interpolation='nearest', cmap='viridis')
plt.colorbar()

plt.title('Mapa de Calor con Interpolaci√≥n Gaussiana')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```
